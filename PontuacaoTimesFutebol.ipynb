```python
def validaPartidas(text):
    while True:
        try:
            valor = int(input(text))
            if valor < 0 or valor > 38:
                print("Digite entre (1 e 38): ")
                continue
            return valor
        except:
            print("Valor inválido!")

#Valida Pontos que só pode ser 0,1,3
def validaPontos (text):
    pontos_possiveis = [0,1,3]
    while True:
        try:
            valor = int(input(text))
            if valor in pontos_possiveis:
                return valor
            else:
                print("Digite apenas 0, 1 ou 3.")
        except:
            print("Valor diferente")
```

```python
#Leitura de Dados:
partidas = validaPartidas("Informe a quantidade de partidas: ")
#Armazena nome do time e pontos
pontos_por_times = dict()

#Laço de pontos por Times

for i in range(partidas):
    time = input(f"Informe nome do time na partida:{i+1}/{partidas}: ").lower().strip() #key
    ponto = validaPontos(f"Informe ponto[0,1,3] da partida {i+1}/{partidas}: ") #values

    if time in pontos_por_times:
        #Verifica se o time já está na lista e soma.
        pontos_por_times[time] += ponto
    else:
        #Time novo, add time e ponto.
        pontos_por_times[time] = ponto

print("Pontuações obtidas!")
```

```python
#Total de Pontos(Somatorio Geral)
total_pontos = 0

for time,ponto in pontos_por_times.items():
    total_pontos += ponto

#Pontos possíveis, no caso de vítoria em todos os jogos:
pontos_possiveis = partidas * 3
#Total de Pontos por Times:

print(f"Quantidade de partidas realizadas: {partidas}")
print(f"Total de pontos: {total_pontos} de {pontos_possiveis} possíveis.")
print("\n")

#Mostra tabela ordenada com o time com mais pontos!
print("TABELA DE TIMES \n")
for index, (time, pontos) in enumerate(sorted(pontos_por_times.items(), key=lambda x: x[1], reverse=True),1):

    print(index, time, pontos)

print("\n")
print("TIME(S) - CAMPEÃO!!!")

#validar o time campeão

lista_campeoes = [] # Cria uma lista de times com maior pontuacao
maior_pontuacao = max(pontos_por_times.values()) #retorna a maior pontuacao

for time,ponto in pontos_por_times.items():
    if ponto == maior_pontuacao:
        lista_campeoes.append(time)
        print(f"O time campeão foi o: {time.upper()} com {ponto} pontos.")
```
